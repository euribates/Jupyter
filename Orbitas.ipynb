```python
version = 'Orbitas 1.0'

import math
import pygame
import random
import time

from pygame.locals import Rect
from math import pi, sin, cos
from vectores import Vector2, zero

SIZE = WIDTH, HEIGHT = 800, 640  # Tama√±o de pantalla

BLACK = (0, 0, 0)                # Colores
WHITE = (255, 255, 255)
SILVER = (172, 172, 172)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
CYAN = (0, 255, 255)
MAGENTA = (255, 0, 255)
YELLOW = (255, 255, 0)

FPS = 30                         # Velocidad del juego
        
def in_game():
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            return False
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_ESCAPE:
                return False
    return True
    
def draw_vector(screen, color, vector, origin=zero):
    pygame.draw.line(screen, color, origin, origin+vector)
    pygame.draw.circle(screen, color, origin, 3)
    arrow_point = origin + vector
    orientation = vector.unit() * 10
    orientation.theta -= 7*pi/8
    left_wing =  arrow_point + orientation 
    orientation.theta -= pi/4
    right_wing =  arrow_point + orientation
    pygame.draw.polygon(screen, color, [arrow_point, left_wing, right_wing])
    
    
    
    
pygame.init()
try:
    pygame.display.set_caption(version)
    screen = pygame.display.set_mode(SIZE, 0, 24)
    limits = pygame.Rect((0, 0), SIZE)
    sun_position = Vector2(*limits.center)
    earth_position = Vector2(200, 0)
    moon_position = Vector2(75, 0)
    clock = pygame.time.Clock()  
    vueltas = 0
    points = []
    while in_game():
        earth_position.theta += pi / 360.
        moon_position.theta += pi / 21.
        screen.fill(BLACK)
        pygame.draw.circle(screen, YELLOW, sun_position, 28)
        pygame.draw.circle(screen, CYAN, sun_position+earth_position, 8)
        pygame.draw.circle(screen, SILVER, sun_position+earth_position+moon_position, 6)
        if vueltas < 180:
            draw_vector(screen, (0, 255, 0, 32), sun_position)
            draw_vector(screen, GREEN, earth_position, sun_position)
            draw_vector(screen, GREEN, moon_position, sun_position+earth_position)
        else:
            points.insert(0, sun_position+earth_position+moon_position)
            if len(points) > 512:
                points = points[0:512]
            for p in points:
                pygame.draw.circle(screen, MAGENTA, p, 2)
            
        pygame.display.update()
        clock.tick(FPS)
        vueltas += 1
finally:
    pygame.quit()
```
