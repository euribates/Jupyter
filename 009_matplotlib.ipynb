```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 100)
y = x**3 - x**2
plt.plot(x, y)
```

```python
x = np.linspace(-2, 2, 50)
y = x**3
plt.subplot(1,3,1)
plt.plot(x, y, 'r--')
plt.subplot(1,3,2)
plt.plot(y, x, 'g*-')
plt.subplot(1,3,3)
plt.plot(x, x**2, 'b*-');
```
