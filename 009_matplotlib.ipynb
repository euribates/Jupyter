## Gr√°ficos con Matplotlib


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(-10, 10, 100)
y = x**3 - x**2
plt.plot(x, y)
```

Podemos indicar distintos colores y s√≠mbolos a utilizar en las gr√°ficas con unas c√≥digos (no demasiado claros, la verdad sea dicha), como en el siguiente ejemplo:

```python
x = np.linspace(-2, 2, 30)
y = x**3
fig = plt.figure()
plt.subplot(1,3,1)
plt.plot(x, y, 'r--')
plt.subplot(1,3,2)
plt.plot(y, x, 'g*-')
plt.subplot(1,3,3)
plt.plot(x, x**2, 'b^-')
fig.set_figwidth(14);
```

Aqui la cadena `r--` se interpreta como _L√≠nea discontinua (--), roja (r)_, la expresi√≥n `g*-` se interpreta como _una l√≠nea verde, continua, y marcando los valores con asteriscos_, y finalmente `b^-` significa _l√≠nea continua azul, marca los valores individuales con un tri√°ngulo_

Algunos de los tipos de l√≠neas que podemos especificar son los siguientes:
    
| C√≥digo | Significado       |
|--------|-------------------|
|  :     | L√≠nea de puntos   |
|  -.    | L√≠nea raya/punto  |
|  --    | L√≠nea discontinua |
|  -     | L√≠nea continua    |

Con respecto a los marcadores, hay much√≠simas posibilidades, en la siguiente tabla mostraremos algunas:
    
| C√≥digo | Significado                        | C√≥digo | Significado      |
|--------|------------------------------------|--------|------------------|
| s      | Cuadrado solido                    | X      | Aspa             |
| 8      | Hex√°gono                           | P      | Cruz             |
| >      | Tri√°ngulo apuntado a la derecha    | d      | Diamante         |  
| <      | Tri√°ngulo apuntado a la izquierda  | D      | Rombo            |
| ^      | Tri√°ngulo apuntado hacia arriba    | H      | Hex√°gono         |
| v      | Tri√°ngulo apuntado hacia abajo     | *      | Estrella         |
| o      | C√≠rculo                            | p      | Pant√°gono        |


```python
x = np.linspace(-2*math.pi, 2*math.pi, 30)
y1 = np.sin(x)
y2 = np.cos(x)
fig = plt.figure()
fig.set_figwidth(15);  # 15 pulgadas
plt.plot(x, y1, 'X--')
plt.plot(x, y2, 'kP:')
```

### Colores

En cuanto a los colores, podemos usar alguno de los colores predefinidocs:
    
| C√≥digo | Color |
|--------|-------|
| b | Az√∫l (Blue) |
| g | Verde (Green) 
| r | Rojo (red) |
| c | Cian (Cyan) |
| m | Magenta |
| y | Amarillo (Yellow) |
| k | Negro (Black) |

Tambien podemos usar un c√≥digo de color html, como en el siguiente caso:

```python
x = np.linspace(-8*math.pi, 8*math.pi, 170)
y = np.sin(x) * x
fig = plt.figure()
fig.set_figwidth(15);  # 15 pulgadas
plt.plot(x, y, ':', color='#5fba7d', marker='X', markersize=10)
```

Obs√©rvese que en el este √∫ltimo ejemplos hemos especificado las distintas caracter√≠sticas del trazo usando pa≈ïametros por nombre (`color`, `marker`, `markersize`), lo que puede ser m√°s sencillo.
