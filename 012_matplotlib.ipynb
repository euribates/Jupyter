## Gráficos con Matplotlib


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(-10, 10, 100)
y = x**3 - x**2
plt.plot(x, y)
```

Podemos indicar distintos colores y símbolos a utilizar en las gráficas con unas códigos (no demasiado claros, la verdad sea dicha), como en el siguiente ejemplo:

```python
x = np.linspace(-2, 2, 30)
y = x**3
fig = plt.figure()
plt.subplot(1,3,1)
plt.plot(x, y, 'r--')
plt.subplot(1,3,2)
plt.plot(y, x, 'g*-')
plt.subplot(1,3,3)
plt.plot(x, x**2, 'b^-')
fig.set_figwidth(14);
```

Aqui la cadena `r--` se interpreta como _Línea discontinua (--), roja (r)_, la expresión `g*-` se interpreta como _una línea verde, continua, y marcando los valores con asteriscos_, y finalmente `b^-` significa _línea continua azul, marca los valores individuales con un triángulo_

Algunos de los tipos de líneas que podemos especificar son los siguientes:
    
| Código | Significado       |
|--------|-------------------|
|  :     | Línea de puntos   |
|  -.    | Línea raya/punto  |
|  --    | Línea discontinua |
|  -     | Línea continua    |

Con respecto a los marcadores, hay muchísimas posibilidades, en la siguiente tabla mostraremos algunas:
    
| Código | Significado                        | Código | Significado      |
|--------|------------------------------------|--------|------------------|
| s      | Cuadrado solido                    | X      | Aspa             |
| 8      | Hexágono                           | P      | Cruz             |
| >      | Triángulo apuntado a la derecha    | d      | Diamante         |  
| <      | Triángulo apuntado a la izquierda  | D      | Rombo            |
| ^      | Triángulo apuntado hacia arriba    | H      | Hexágono         |
| v      | Triángulo apuntado hacia abajo     | *      | Estrella         |
| o      | Círculo                            | p      | Pantágono        |


```python
x = np.linspace(-2*math.pi, 2*math.pi, 30)
y1 = np.sin(x)
y2 = np.cos(x)
fig = plt.figure()
fig.set_figwidth(15);  # 15 pulgadas
plt.plot(x, y1, 'X--')
plt.plot(x, y2, 'kP:')
```

### Colores

En cuanto a los colores, podemos usar alguno de los colores predefinidocs:
    
| Código | Color |
|--------|-------|
| b | Azúl (Blue) |
| g | Verde (Green) 
| r | Rojo (red) |
| c | Cian (Cyan) |
| m | Magenta |
| y | Amarillo (Yellow) |
| k | Negro (Black) |

Tambien podemos usar un código de color html, como en el siguiente caso:

```python
x = np.linspace(-8*math.pi, 8*math.pi, 170)
y = np.sin(x) * x
fig = plt.figure()
fig.set_figwidth(15);  # 15 pulgadas
plt.plot(x, y, ':', color='#5fba7d', marker='X', markersize=10)
```

Obsérvese que en el este último ejemplos hemos especificado las distintas características del trazo usando paŕametros por nombre (`color`, `marker`, `markersize`), lo que puede ser más sencillo.
