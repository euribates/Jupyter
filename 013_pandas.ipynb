### Copiar un dataframe

Podemos copiar un dataframe para manipularlo, mantenido ibntacto el dataframe original, para eso se usa el mtodo `copy`.

```python
df = pokemon.copy()
```

### Aadir una columna al dataframe

Podemos asignar nuevas columnas a un data frame, simplemente asignandole a un nuevo nombre. Normalmente los datos se obtienen como combinacin de otras columnas, funciones, etc...

```python
df['aprox_density'] = df['weight'] / df['height']
df.head()
```

```python
df.loc[:, ['id', 'aprox_density']].plot()
```

```python
df[df.aprox_density == df.aprox_density.max()]
```
