## Starfield

```python
import math
import random
import time
import pygame
from pygame.locals import *
import vectores
from vectores import Vector2


# Constantes

SIZE = WIDTH, HEIGHT = 800, 640 # Tamaño de pantalla
LIMITS = pygame.Rect((0, 0), (WIDTH, HEIGHT))

BLACK = (0, 0, 0)               # Colores
WHITE = (255, 255, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
CYAN = (0, 255, 255)
MAGENTA = (255, 0, 255)
YELLOW = (255, 255, 0)

FPS = 30                       # Velocidad del juego

class Star:
    
    orientation = vectores.right
    
    def __init__(self):
        self.pos = Vector2(
            random.randrange(0, WIDTH),
            random.randrange(0, HEIGHT),
            )
        self.color = (
            random.randrange(128, 255),
            random.randrange(128, 255),
            random.randrange(128, 255),
            )
        self.speed = random.randrange(1, 6)
        self.size = random.randrange(1, 4)
        
        
    def update(self):
        self.pos += Star.orientation * self.speed
        if self.pos.x < 0:
            self.pos.x += WIDTH
        elif self.pos.x > WIDTH:
            self.pos.x -= WIDTH
        if self.pos.y < 0:
            self.pos.y += HEIGHT
        elif self.pos.y > HEIGHT:
            self.pos.y -= HEIGHT
            
        
    def draw(self, surface):
        pygame.draw.circle(surface, self.color, self.pos, self.size)
        
def play_game():
    pygame.init()
    try:
        pygame.display.set_caption("Starfield full")
        screen = pygame.display.set_mode(SIZE, 0, 24)
        # Parte de inicialización del juego
        stars = [Star() for _ in range(50)]
        clock = pygame.time.Clock()    
        in_game = True
        while in_game:
        # Obtener datos de entrada
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    in_game = False
                elif event.type == KEYDOWN:
                    print(event.type, event.key)
                    if event.key == K_w:
                        Star.orientation = vectores.up
                    elif event.key==K_a:
                        Star.orientation = vectores.left
                    elif event.key==K_s:
                        Star.orientation = vectores.down
                    elif event.key==K_d:
                        Star.orientation = vectores.right
                    elif event.key==K_SPACE:
                        Star.orientation = vectores.zero

            # Recalcular el estado del juego, en base al estado actual y a las entradas
            for star in stars:
                star.update()  # pasamos la responsabilidad de actualizar su estado a la estrella
            # Representamos el nuevo estado
            screen.fill(BLACK)
            for star in stars:
                star.draw(screen)  # pasamos la responsabilidad de dibujarse a la estrella
            pygame.display.update()
            clock.tick(FPS)
    finally:
        pygame.quit()
        
play_game()
```
